* *Resumo:* Este trabalho tem como objetivo cria um menu interativo de uma calculadora onde se pode selecionar as opções desejadas como a soma dos números numa lista, o cálculo da sua média, a indicação do maior ou menor número da lista, a verificação da ordem crescente ou descrescente da lista bem como a procura de um número específico na lista digitada. 

```python
import random

# Variável interna para guardar a lista de números
lista_numeros = []

def criar_lista_aleatoria(): #  criar uma lista de números aleatórios entre 1 e 100 que será guardada na variável interna
    global lista_numeros
    lista_numeros = [random.randint(1, 100) for _ in range(10)]
    print("Lista de números aleatórios criada.")

def criar_lista_utilizador(): # criar uma lista com números introduzidos pelo utilizador, que será guardada na variável interna
    global lista_numeros 
    lista_numeros = []
    n = int(input("Quantos números deseja inserir na lista? "))
    for _ in range(n):
        numero = int(input("Insira um número: "))
        lista_numeros.append(numero)
    print("Lista de números inseridos criada.")

# Nestas primeiras opções, se a variável interna já tiver uma lista, esta será sobreposta/apagada pela nova lista;

def calcular_soma(): # calcula a soma dos elementos na lista no momento
    soma = sum(lista_numeros)
    print(f"Soma dos elementos na lista: {soma}")

def calcular_media(): # calcula a média dos elementos na lista no momento 
    media = sum(lista_numeros) / len(lista_numeros)
    print(f"Média dos elementos na lista: {media:.2f}")
    
def encontrar_maior(): # calcula o maior elemento da lista no momento
    maior = max(lista_numeros)
    print(f"Maior elemento na lista: {maior}")

def encontrar_menor(): # calcula o menor elemento da lista no momento;
    menor = min(lista_numeros)
    print(f"Menor elemento na lista: {menor}")

def verificar_ordem_crescente(): # aplicação indica (Sim/Não) se a lista está ordenada por ordem crescente
    if lista_numeros == sorted(lista_numeros):
        print("A lista está ordenada por ordem crescente: Sim")
    else:
        print("A lista está ordenada por ordem crescente: Não")

def verificar_ordem_decrescente(): # aplicação indica (Sim/Não) se a lista está ordenada por ordem decrescente
    if lista_numeros == sorted(lista_numeros, reverse=True):
        print("A lista está ordenada por ordem decrescente: Sim")
    else:
        print("A lista está ordenada por ordem decrescente: Não")

def procurar_elemento(): # procura um elemento na lista, se o encontrar devolve a sua posição, devolve -1 se o elemento não estiver na lista;
    elemento = int(input("Digite o número que procura: "))
    if elemento in lista_numeros:
        posicao = lista_numeros.index(elemento)
        print(f"O elemento {elemento} está na posição {posicao} da lista.")
    else:
        print("-1")

def main():
    while True:
        print("\nMenu:")
        print("1. Criar uma Lista Aleatória")
        print("2. Criar uma Lista do Utilizador")
        print("3. Soma da Lista")
        print("4. Média da Lista")
        print("5. Maior da Lista")
        print("6. Menor da Lista")
        print("7. Verificar se a Lista está ordenada por ordem crescente")
        print("8. Verificar se a Lista está ordenada por ordem decrescente")
        print("9. Procurar se um elemento está na Lista")
        print("0. Sair") #  Se o utilizador selecionar a opção 0, a aplicação termina mostrando a lista que está nesse momento guardada.

        opcao = input("Escolha uma das opções: ")

        if opcao == "1":
            criar_lista_aleatoria()
        elif opcao == "2":
            criar_lista_utilizador()
        elif opcao == "3":
            calcular_soma()
        elif opcao == "4":
            calcular_media()
        elif opcao == "5":
            encontrar_maior()
        elif opcao == "6":
            encontrar_menor()
        elif opcao == "7":
            verificar_ordem_crescente()
        elif opcao == "8":
            verificar_ordem_decrescente()
        elif opcao == "9":
            procurar_elemento()
        elif opcao == "0":
            print(f"Lista atual: {lista_numeros}")
            print("Saiu do programa. Adeus!")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
