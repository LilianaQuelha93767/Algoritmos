* *Resumo:* Este trabalho tem como objetivo criar uma aplicação que utiliza uma série de operações trabalhando só com frações.

```python
import math
n1 = int(input("Insira o primeiro numerador: 1\n"))
d1 = int(input("Insira o primeiro denominador: 1\n"))
n2 = int(input("Insira o segundo numerador: 2\n"))
d2 = int(input("Insira o segundo denominador: 2\n"))
operacao = input("1 - Somar\n2 - Subtrair\n3 - Multiplicar\n4 - Dividir\n0 - Sair\n")

if(operacao == "1"):
    # se possui mesmo denominador
    if(d1 == d2):
        numerador = n1+n2
        denominador = d1
        
    # se possui um denominador diferente
    else:
        numerador = (n1*d2)+(n2*d1)
        denominador = d1*d2
        
    # se possui um divisor comum
    div_comum = math.gcd(numerador,denominador)
    numerador = numerador//div_comum
    denominador = denominador//div_comum

    # Resultado da fração final
    fracao = str(numerador)+"/"+str(denominador)
    print("Escolhendo a Operação de Adição, o resultado final é: "+fracao)

elif(operacao == "2"):
    # se possui mesmo denominador
    if(d1 == d2):
        numerador = n1-n2
        denominador = d1
        
    # se possui um denominador diferente
    else:
        numerador = (n1*d2)-(n2*d1)
        denominador = d1*d2
        
    # se possui um divisor comum
    div_comum = math.gcd(numerador,denominador)
    numerador = numerador//div_comum
    denominador = denominador//div_comum

    # Resultado da fração final
    fracao = str(numerador)+"/"+str(denominador)
    print("Escolhendo a Operação de Subtraçao, o resultado final é: "+fracao)

elif (operacao == "3"):
    numerador = n1*n2
    denominador = d1*d2

    # se possui um divisor comum
    div_comum = math.gcd(numerador,denominador)
    numerador = numerador//div_comum
    denominador = denominador//div_comum

    # Resultado da fração final
    fracao = str(numerador)+"/"+str(denominador)
    print("Escolhendo a Operação de Multiplicação, o resultado final é: "+fracao)

elif (operacao == "4"):
    numerador = n1*d2
    denominador = n2*d1

    # se possui um divisor comum
    div_comum = math.gcd(numerador,denominador)
    numerador = numerador//div_comum
    denominador = denominador//div_comum

    # Resultado da fração final
    fracao = str(numerador)+"/"+str(denominador)
    print("Escolhendo a Operação de Divisão, o resultado final é: "+fracao)

else:
    print("Operação Interrompida")
```
